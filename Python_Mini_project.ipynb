# Iris Dataset Cleaning , Evaluation and Visualization



```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_openml
%matplotlib inline
# Loading the Iris dataset from UCI Machine Learning Repository
iris = fetch_openml(name='iris', version=1, as_frame=True)
iris_df = iris.frame
```

```python
print("Original Iris Dataset , first few rows:")
iris_df.head()
```

```python
iris_df.columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species']

# Checking for missing values
print("Missing Values:")
print(iris_df.isnull().sum())
```

```python
# Data Visualization: Bar plot of Species Count
species_count = iris_df['species'].value_counts()

plt.figure(figsize=(8, 6))
plt.bar(species_count.index, species_count, color=['skyblue', 'salmon', 'lightgreen'])
plt.title('Count of Each Species')
plt.xlabel('Species')
plt.ylabel('Count')
plt.show()
```

# Inference:
* The bar plot of species count provides a simple overview of the distribution of different species in the iris dataset.
* Each species (Setosa, Versicolor, Virginica) has an equal count in this dataset.
* Basically , one can say , this dataset is balanced.

```python
# Data Visualization: Histogram of Petal Length
plt.figure(figsize=(8, 6))
plt.hist(iris_df['petal_length'], bins=20, color='skyblue', edgecolor='black')
plt.title('Histogram of Petal Length')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Frequency')
plt.show()
```

#Inference:
*   The histogram of petal length shows that most of the flowers have petal lengths between approximately 1.0 cm to 1.7 cm.
*   There is a clear separation of flowers into different bins, suggesting the presence of distinct groups or species based on petal length.



```python
# Data Visualization: Scatter plot of Sepal Length vs Sepal Width
plt.figure(figsize=(8, 6))
plt.scatter(iris_df['sepal_length'], iris_df['sepal_width'], c=iris_df['species'].astype('category').cat.codes, cmap='viridis', edgecolors='k')
plt.title('Scatter Plot of Sepal Length vs Sepal Width')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.show()
```


# Inference:
* The scatter plot of sepal length vs sepal width provides insights into the distribution of these two features.
* There is some overlap between the species, but Setosa tends to have shorter and wider sepals compared to Versicolor and Virginica.

```python
# Data Visualization: Bar plot of Mean Petal Length for each Species
mean_petal_length = iris_df.groupby('species')['petal_length'].mean()
plt.figure(figsize=(8, 6))
mean_petal_length.plot(kind='bar', color=['skyblue', 'salmon', 'lightgreen'])
plt.title('Mean Petal Length for each Species')
plt.xlabel('Species')
plt.ylabel('Mean Petal Length (cm)')
plt.show()
```

# Inference:
* The bar plot of mean petal length for each species visually compares the average petal length.
* Virginica tends to have the longest petals on average, followed by Versicolor and Setosa.

```python
# Assuming 'iris_df' is your DataFrame containing the iris dataset
mean_petal_width = iris_df.groupby('species')['petal_width'].mean()
plt.figure(figsize=(8, 6))
plt.bar(mean_petal_width.index, mean_petal_width, color=['grey', 'blue', 'red'])
plt.title('Mean Petal Width for each Species')
plt.xlabel('Species')
plt.ylabel('Mean Petal Width (cm)')
plt.show()
```

# Inference:
* The bar plot of mean petal width for each species visually compares the average petal width.
* Virginica tends to have the longest petals on average, followed by Versicolor and Setosa.
